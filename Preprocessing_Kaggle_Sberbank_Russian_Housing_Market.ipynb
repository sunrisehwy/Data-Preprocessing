**Dataset Setup**

```
#Check the Column name
trainCols = list(trainData.columns.values)
testCols = list(testData.columns.values)
print(trainCols)
print(testCols)
```

```
#Extract Target Column before doing missing value substitutions 
#Seperate Target column from Dataset
trainData = trainData.iloc[:, :-1] #extracting training data without the target column
y_train = train["price_doc"] #make copy of target column
```

```
print(trainData.shape)
print(testData.shape)
```

```
#Check the Column name
trainCols = list(trainData.columns.values)
testCols = list(testData.columns.values)
print(trainCols)
print(testCols)
```

```
#drop.id_coloumn
trainData=trainData.drop('id', axis= 1)
testData=testData.drop('id', axis= 1)
```

**Handling Missing Data - train Data, test Data**

**(1) DROP COLUMNS**

```
#Combine Train data and test data first so that the SAME COLUMNS are DROPPED in each

combined_Data = pd.concat([trainData, testData], keys=[0,1])
combined_Data.isnull().sum().sort_values(ascending=False)
```

```
#Checking the Percentage of Null in category

percent = combined_Data.isnull().sum() / len(combined_Data) * 100
null_percent_df = pd.DataFrame({'column': combined_Data.columns,
                                 'pct_null': percent}
                                )
null_percent_df.sort_values(by='pct_null',inplace=True, ascending=False)
null_percent_df.head(150)
```

```
#Define threshold for dropping columns
threshold=int(0.60*(combined_Data.shape[0]))
print(threshold)

#Drop columns that have less than "thresh" number of non_Nans
td1=combined_Data.dropna(thresh=threshold,axis=1)

print('---combine_data shape---')
print (combined_Data.shape)
print('---remaining_data shape---')
print(td1.shape)
```

```
#look at what other columns have missing values
#Checking the Percentage of Null in category

percent_tdl = td1.isnull().sum() / len(td1) * 100
null_percent_df_td1 = pd.DataFrame({'column': td1.columns,
                                 'pct_null': percent_tdl}
                                )
null_percent_df_td1.sort_values(by='pct_null',inplace=True, ascending=False)
null_percent_df_td1.head(150)
```

**(2) IMPUTE MISSING VALUES**

```
#Impute MEAN values for NaN in numeric columns 
numeric=td1.select_dtypes(include=['int','float64']).columns
for num in numeric:
  td1[num]=td1[num].fillna(td1[num].mean())

#Impute MODE values for NaN in categorical Columns
train_cat_cols = td1.select_dtypes(exclude=['int','float64']).columns #selecting the categorical columns
for colss in train_cat_cols:
  if(td1.iloc[0][colss]=="N"): 
        td1[colss]=td1[colss].fillna("N")
  else:
    td1[colss]=td1[colss].fillna(td1[colss].mode())
  
print(td1.head(20))
```

```
#Check remaining missing values 
td1.isnull().sum().sort_values(ascending=False)
```

```
#Check the data type of column with remaining missing value

td1['product_type'].unique()
```

```
#Fill Null to 'NA' 
td1['product_type'] = td1['product_type'].fillna('NA')
td1['product_type'].unique()
```

**Handling Missing Data - macro Data**

**(1) Drop Columns**

```
#Checking the Percentage of Null in category

percent_macroData = macroData.isnull().sum() / len(macroData) * 100
null_percent_df_macroData = pd.DataFrame({'column': macroData.columns,
                                 'pct_null': percent_macroData}
                                )
null_percent_df_macroData.sort_values(by='pct_null',inplace=True, ascending=False)
null_percent_df_macroData.head(150)
```

```
#Define threshold for dropping columns
threshold_macro=int(0.95*(macroData.shape[0]))
print(threshold_macro)

#Drop columns that have less than "thresh" number of non_Nans
td2=macroData.dropna(thresh=threshold_macro,axis=1)

print('---combine_data shape---')
print(macroData.shape)
print('---remaining_data shape---')
print(td2.shape)
```

```
#Checking the Percentage of Null in category

percent_td2 = td2.isnull().sum() / len(td2) * 100
null_percent_df_td2 = pd.DataFrame({'column': td2.columns,
                                 'pct_null': percent_td2}
                                )
null_percent_df_td2.sort_values(by='pct_null',inplace=True, ascending=False)
null_percent_df_td2.head(150)
```

**(2) Impute missing values**

```
#Impute MEAN values for NaN in numeric Columns 
numeric=td2.select_dtypes(include=['int','float64']).columns
for num in numeric:
  td2[num]=td2[num].fillna(td2[num].mean())

#Impute MODE values for NaN in categoricla columns 
train_cat_cols = td2.select_dtypes(exclude=['int','float64']).columns #selecting the categorical columns
for colss in train_cat_cols:
  if(td2.iloc[0][colss]=="NA"): 
        td2[colss]=td2[colss].fillna("NA")
  else:
    td2[colss]=td2[colss].fillna(td2[colss].mode())
  
print(td2.head(20))
```

```
#Check any Remaining missing values left  
td2.isnull().sum().sort_values(ascending=False)
```

**Merging macro data, train data, and test data set**

```
#merge_macro_data_set with train, test Data 
td_f = pd.merge(td1, td2, how="left", on="timestamp")
td_f.shape
```

**Feature Engineering for 'timestamp' feature**

```
#Extract the year/month/day features from a ‘timestamp’ 

td_f['timestamp'] = pd.DatetimeIndex(td_f.timestamp)
td_f['year'] = pd.DatetimeIndex(td_f['timestamp']).year
td_f['month'] = pd.DatetimeIndex(td_f['timestamp']).month
td_f['day'] = pd.DatetimeIndex(td_f['timestamp']).day

td_f.head()
```

```
#drop the previous timestamp column 
td_ff=td_f.drop('timestamp', axis= 1)
```

```
#converting'month' and 'day' to categorical data
td_ff['month']= td_ff['month'].astype('str')
td_ff['day']= td_ff['day'].astype('str')
td_ff.dtypes
td_ff.shape
```

**Data Scaling**

```
#Separate Train data and test data for data scaling 

trainData = td_ff.iloc[:30471]
testData= td_ff.iloc[30471:]

print(trainData.shape)
print(testData.shape)
```

```
#trainData_Scaling
scal = MinMaxScaler()
num_col_train = trainData._get_numeric_data().columns.tolist()
trainData[num_col_train] = scal.fit_transform(trainData[num_col_train])

#testData_Scaling
num_col_test = testData._get_numeric_data().columns.tolist()
testData[num_col_test] = scal.transform(testData[num_col_test])
```

**One Hot encoding**




```
#Concatenating
combined_Data = pd.concat([trainData, testData], keys=[0,1])
combined_Data.head()
```

```
#The below function returns a list of categorical features which are not numeric. 
train_cat_cols = combined_Data.select_dtypes(exclude=['float','int']).columns #selecting the categorical columns
print(train_cat_cols.shape)
print(train_cat_cols)

combined_Data = pd.get_dummies(combined_Data,train_cat_cols)
combined_Data.head(10)
```

```
#Separate Train data and test data
trainData = combined_Data.xs(0)
testData = combined_Data.xs(1)
print(trainData.shape)
print(testData.shape)
```

**Trials**

```
X_train=trainData
X_test=testData
```

```
from sklearn.tree import DecisionTreeRegressor

df = DecisionTreeRegressor(max_depth=10)
df.fit(X_train, y_train)

pred = df.predict(X_test)
output = pd.DataFrame({
    'id': test['id'],
    'price_doc': pred
})
output.to_csv(r'/content/drive/MyDrive/CIS508/team/output_train_test_60_macro_95.csv', index=False)
```
